# Medindo qualidade da água

A poluição dos recursos hídricos, além de prejudicar o equilíbrio ecológico dos ecossistemas, interfere diretamente na qualidade da água disponível para o consumo humano. 

Por intuição, achamos que a água boa para o consumo deve ser cristalina, sem gosto e sem odor. Entretanto, somente essas características não garantem a boa qualidade desse líquido tão precioso para a vida.

Contaminações por vírus, bactérias e substâncias tóxicas microscópicas são imperceptíveis a olho nu e podem prejudicar a nossa saúde. Por isso, além de não ter sabor, cheiro e cor, as águas consumíveis devem estar enquadradas em parâmetros mais específicos de qualidade.



## Informação sobre as colunas

1. pH:
PH  é um parâmetro importante na avaliação do balanço ácido-base na água. É também o indicador de condições ácidas ou alcalinas. A OMS recomenda um limite máximo permissível de pH, entre 6.5 e 8.5. <br/><br/>

2. Dureza:
A dureza é causada principalmente por sais de magnésio e cálcio. Esses sais são dissolvidos de depósitos geológicos por onde a água passa. O período de tempo que a água está em contato com o material de produção de dureza ajuda a determinar quanta dureza existe na água pura. <br/><br/>

3. Sólidos (Total Dissolved Solids - TDS):
A água tem a habilidade de dissolver um grande número de minerais inorgânicos e orgânicos como potássio, cálcio, sódio, bicarbonatos, cloretos, magnésio, sulfatos etc. Esses minerais produzem cores e sabores indesejados na água. Este é um parâmetro importante para o uso de água. A água com alto valor TDS indica que a água é altamente mineralizada. Limites aceitáveis para o TDS são entre 500 mg/l 1000 mg/l, prescrito para beber. <br/><br/>

4. Cloraminas:
Cloro e cloraminas são os principais desinfetantes usados em sistemas públicos de água. Cloraminas são comumente formadas quando a amônia é adicionada ao cloro, para tratar a água para beber. Níveis de cloro que permanecem em 4 miligramas por litro (mg/L ou 4 partes por milhão (ppm)) são consideráveis seguras para beber. <br/><br/>

5. Sulfato:
Sulfatos são substâncias que surgem naturalmente em minerais, pedras e no solo. Eles estão presentes no ar ambiente, na água subterrânea, nas plantas e na comida. O principal uso comercial do sulfato é nas indústrias químicas. A concentração de sulfato na água do mar é em torno de  2700 miligramas por litro (mg/L). Varia de 3 para 30 mg/L  na maioria das fontes de água fresca, contudo concentrações bem maiores (1000 mg/L) são encontradas em alguns lugares. <br/><br/>

6. Condutividade:
Água pura não é um bom condutor de eletricidade, em vez disso é um bom isolante. O aumento da concentração de íons melhora a condutividade elétrica (CE) da água. No geral, o aumento de sólidos dissolvidos na água determina a CE. A CE na verdade mede o processo iônico de uma solução que permite a transmissão da corrente. De acordo com as normas da OMS, a CE não pode exceder 400 μS/cm. <br/><br/>

7. Carbono orgânico:
O Carbono Orgâncio Total (TOC) em fontes de água vem da degradação da matéria orgânica natural bem como de fontes sintéticas. O TOC é uma medida da quantidade total de carbono em compostos orgânicos na água pura. De acordo com a Agência de Proteção Ambiental dos Estados Unidos, o nível de TOC em água tratada/para beber tem que ser menor do que 2 mg/L, e menor que 4 mg/L na fonte de água usada para tratamento. <br/><br/>

8. Trihalometanos (THM):
THMs são químicos que podem ser encontrados em água tratada com cloro. A concentração em THMs em água para beber varia de acorda com o nível de material orgânico na água, a quantidade de cloro necessária para tratar a água, e a temperatura da água que está sendo tratada. Níveis THM em torno de 80 ppm é considerada segura em água para beber. <br/><br/>

9. Turbidez:
A turbidez da água depende da qualidade de matéria sólida presente no estado suspenso. É uma medida das propriedades de emissão de luz e o teste é utilizado para indicar a qualidade da descarga de resíduos em relação à matéria coloidal. O valor recomendado pela OMS é 5.00 NTU. <br/><br/>

10. Potabilidade:
Indica se a água é segura para consumo humano onde 1 significa potável e 0 não potável.

### 2.3 Box Plots

```python
cols = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity',
       'Organic_carbon', 'Trihalomethanes', 'Turbidity']
```

```python
from tqdm import tqdm_notebook

def boxPlotter(dataset, columnName):
    #Plota boxplots pela coluna dada como parâmetro
    sns.catplot(x="Potability", y=columnName, data=ndf, kind="box");
for column in tqdm_notebook(cols, desc = "Your Charts are being ready"):
    boxPlotter(ndf, column)
```

```python
def pieChartPlotter(dataset, columnName):
    values = dataset[columnName].value_counts()
    labels = dataset[columnName].unique()
    pie, ax = plt.subplots(figsize=[10, 6])

    patches, texts, autotexts = ax.pie(values, labels=labels, autopct='%1.2f%%', shadow=True, pctdistance=.5,explode=[0.06]*dataset[columnName].unique()
                                       )

    plt.legend(patches, labels, loc="best")
    plt.title(columnName, color='white', fontsize=14)
    plt.setp(texts, color='white', fontsize=20)
    plt.setp(autotexts, size=10, color='black')
    autotexts[1].set_color('black')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

```python
pieChartPlotter(ndf, 'Potability') 
```
